```python
# Ankara bilinen ilçeleri
ilceler = [
    'Akyurt', 'Altındağ', 'Ayaş', 'Bala', 'Beypazarı', 'Çamlıdere',
    'Çankaya', 'Çubuk', 'Elmadağ', 'Etimesgut', 'Evren', 'Gölbaşı',
    'Güdül', 'Haymana', 'Kahramankazan', 'Kalecik', 'Keçiören',
    'Kızılcahamam', 'Mamak', 'Nallıhan', 'Polatlı', 'Pursaklar',
    'Sincan', 'Şereflikoçhisar', 'Yenimahalle'
]

# İlçe ve mahalle bilgilerini tutan sözlükler - mahalle adına göre ilçe belirlemek için
mahalle_ilce_eslestirme = {
    # Keçiören mahalleleri
    'Sanatoryum': 'Keçiören',
    'Etlik': 'Keçiören',
    'İncirli': 'Keçiören',
    'Esertepe': 'Keçiören',
    'Yayla': 'Keçiören',
    'Ufuktepe': 'Keçiören',
    'Aktepe': 'Keçiören',
    'Kalaba': 'Keçiören',
    'Şenlik': 'Keçiören',
    'Güçlükaya': 'Keçiören',
    'Bağlum': 'Keçiören',

    # Etimesgut mahalleleri
    'Eryaman': 'Etimesgut',
    'Çarşı': 'Etimesgut',
    'Merkez': 'Etimesgut',

    # Çankaya mahalleleri
    'Öveçler': 'Çankaya',
    'Kırkkonaklar': 'Çankaya',
    'Gaziosmanpaşa': 'Çankaya',
    'Gop': 'Çankaya',
    'Kızılay': 'Çankaya',
    'Bahçelievler': 'Çankaya',
    'Çayyolu': 'Çankaya',
    'Ümitköy': 'Çankaya',
    'Ayrancı': 'Çankaya',
    'Dikmen': 'Çankaya',
    'Yıldız': 'Çankaya',
    'Cebeci': 'Çankaya',
    'Sıhhiye': 'Çankaya',
    'Zafer': 'Çankaya',
    'Anıttepe': 'Çankaya',
    'Balgat': 'Çankaya',
    'Küçükesat': 'Çankaya',
    'Zafertepe': 'Çankaya',
    'Tunalı Hilmi': 'Çankaya',
    'Bilkent': 'Çankaya',
    'Bilkentplaza': 'Çankaya',
    'Beytepe': 'Çankaya',
    'Kavaklıdere': 'Çankaya',
    'Hürriyet': 'Çankaya',

    # Yenimahalle mahalleleri
    'Demetevler': 'Yenimahalle',
    'Batıkent': 'Yenimahalle',
    'A.O.Çiftliği': 'Yenimahalle',
    'Aoç': 'Yenimahalle',
    'Ostim': 'Yenimahalle',
    'Şentepe': 'Yenimahalle',
    'İstasyon': 'Yenimahalle',
    'Karşıyaka': 'Yenimahalle',
}

konum_sutunu = None
if 'Neighborhood' in sahininden_veri.columns:
    konum_sutunu = 'Neighborhood'
elif 'Konum' in sahininden_veri.columns:
    konum_sutunu = 'Konum'
else:
    raise ValueError("DataFrame'de 'Neighborhood' veya 'Konum' sütunu bulunamadı!")

def ayristir(konum):
    if not isinstance(konum, str):
        return None, str(konum)


    for ilce in ilceler:
        if konum.startswith(ilce):
            mahalle = konum[len(ilce):]

            if not mahalle or mahalle.isspace():
                mahalle = "Merkez"
            return ilce, mahalle

    if "Mah." in konum:
        mahalle_adi = konum.split("Mah.")[0].strip()
        # Mahalle adından ilçe belirle
        for anahtar, ilce_adi in mahalle_ilce_eslestirme.items():
            if anahtar in mahalle_adi:
                return ilce_adi, konum

    return None, konum


if 'ILCE' not in sahininden_veri.columns or 'Mahalle' not in sahininden_veri.columns:

    sonuclar = sahininden_veri[konum_sutunu].apply(ayristir)
    sahininden_veri['ILCE'] = sonuclar.apply(lambda x: x[0])
    sahininden_veri['Mahalle'] = sonuclar.apply(lambda x: x[1])
else:

    temp_sonuclar = sahininden_veri[konum_sutunu].apply(ayristir)
    sahininden_veri['ILCE'] = temp_sonuclar.apply(lambda x: x[0])
    sahininden_veri['Mahalle'] = temp_sonuclar.apply(lambda x: x[1])

if 'Konum' in sahininden_veri.columns:
    sahininden_veri['Konum'] = sahininden_veri['Konum'].astype(str).str.replace('\r', '')
sahininden_veri['Mahalle'] = sahininden_veri['Mahalle'].astype(str).str.replace('\r', '')

# Sonuçları görüntüle
sahininden_veri.head()
```

```python
sahininden_veri=sahininden_veri.drop("Neighborhood", axis=1)
sahininden_veri=sahininden_veri.rename(columns={"Mahalle":"Neighborhood"})
```
